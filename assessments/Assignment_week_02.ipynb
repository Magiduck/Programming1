Create a plot with the original data and the interpolated data. An example of such a plot is given below.

```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.layouts import gridplot
from bokeh.io import output_notebook
from bokeh.plotting import ColumnDataSource
output_notebook()
```

```python
p = figure(title='Original vs interpolated Glucose',x_axis_type='datetime', background_fill_color="#fafafa")

###########################################
p.circle(df.index, df.glucose, color='red')
p.line(df.index, df.interpolated, color='blue')
###########################################

p.xaxis.axis_label = 'time'
p.yaxis.axis_label = 'signal'
p.grid.grid_line_color="white"

show(p)
```
